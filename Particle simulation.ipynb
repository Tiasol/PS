```python
import numpy as np
import numpy
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def initial_state(xsize,ysize,particle_number):
    a=np.zeros((ysize,xsize),dtype=int)
    for i in range(particle_number):
        x_position=np.random.randint(xsize)
        y_position=np.random.randint(ysize)
        a[y_position,x_position]=a[y_position,x_position]+1
       
    return a
```

```python
def initial_operation(state):
    initial_operation=[]
    initial_state=np.copy(state)
    (ysize,xsize)=np.shape(initial_state)
    for j in range(ysize):
        for i in range(xsize):
            if initial_state[j,i]==1:
                dire=np.random.randint(4)
                initial_operation.append((1,(j,i),dire))

            elif initial_state[j,i]==3:
                dire=np.random.randint(4)
                initial_operation.append((3,(j,i),dire))

            elif initial_state[j,i]>=4:
                initial_operation.append((4,(j,i)))
    
    return initial_operation 
```

```python
def roll(state,position,direction):
    a=np.copy(state)
    (ysize,xsize)=np.shape(a)
    (y,x)=position
    a[y,x]=a[y,x]-1
    if direction == 0:
        a[(y+1)%ysize,x]=a[(y+1)%ysize,x]+1
        newposition=((y+1)%ysize,x)
        
    elif direction == 1:
        a[y,(x+1)%xsize]=a[y,(x+1)%xsize]+1
        newposition=(y,(x+1)%xsize)
    
    if direction == 2:
        a[(y-1+ysize)%ysize,x]=a[(y-1+ysize)%ysize,x]+1
        newposition=((y-1+ysize)%ysize,x)
    
    if direction == 3:
        a[y,(x-1+xsize)%xsize]=a[y,(x-1+xsize)%xsize]+1
        newposition=(y,(x-1+xsize)%xsize)
    
    b=np.copy(a)
    return (b,newposition)      
```

```python
def explosion2(state,position,direction):
    a=np.copy(state)
    (ysize,xsize)=np.shape(a)
    (y,x)=position
    newposition=[]
    
    a[y,x]=a[y,x]-2
    if direction==0:
        a[(y+1)%ysize,x]=a[(y+1)%ysize,x]+1
        newposition.append(((y+1)%ysize,x))
        a[y,(x+1)%xsize]=a[y,(x+1)%xsize]+1
        newposition.append((y,(x+1)%xsize))
        
    if direction==1:
        a[y,(x+1)%xsize]=a[y,(x+1)%xsize]+1
        newposition.append((y,(x+1)%xsize))
        a[(y-1+ysize)%ysize,x]=a[(y-1+ysize)%ysize,x]+1
        newposition.append(((y-1+ysize)%ysize,x))
    
    if direction==2:
        a[(y-1+ysize)%ysize,x]=a[(y-1+ysize)%ysize,x]+1
        newposition.append(((y-1+ysize)%ysize,x))
        a[y,(x-1+xsize)%xsize]=a[y,(x-1+xsize)%xsize]+1
        newposition.append((y,(x-1+xsize)%xsize))
                        
    if direction==3:
        a[y,(x+1)%xsize]=a[y,(x+1)%xsize]+1
        newposition.append((y,(x+1)%xsize))
        a[(y+1)%ysize,x]=a[(y+1)%ysize,x]+1
        newposition.append(((y+1)%ysize,x))
        
    if direction==4:
        a[(y+1)%ysize,x]=a[(y+1)%ysize,x]+1
        newposition.append(((y+1)%ysize,x))
        a[(y-1+ysize)%ysize,x]=a[(y-1+ysize)%ysize,x]+1
        newposition.append(((y-1+ysize)%ysize,x))
    
    if direction==5:
        a[y,(x+1)%xsize]=a[y,(x+1)%xsize]+1
        newposition.append((y,(x+1)%xsize))
        a[y,(x-1+xsize)%xsize]=a[y,(x-1+xsize)%xsize]+1
        newposition.append((y,(x-1+xsize)%xsize))
        
    b=np.copy(a)
    return (b,newposition)     
```

```python
def explosion3(state,position,direction):
    a=np.copy(state)
    (ysize,xsize)=np.shape(a)
    (y,x)=position
    newposition=[]
    
    a[y,x]=a[y,x]-3
    if direction==0:
        a[y,(x+1)%xsize]=a[y,(x+1)%xsize]+1
        newposition.append((y,(x+1)%xsize))
        a[(y-1+ysize)%ysize,x]=a[(y-1+ysize)%ysize,x]+1
        newposition.append(((y-1+ysize)%ysize,x))
        a[y,(x-1+xsize)%xsize]=a[y,(x-1+xsize)%xsize]+1
        newposition.append((y,(x-1+xsize)%xsize))
        
    if direction==1:
        a[(y-1+ysize)%ysize,x]=a[(y-1+ysize)%ysize,x]+1
        newposition.append(((y-1+ysize)%ysize,x))
        a[y,(x-1+xsize)%xsize]=a[y,(x-1+xsize)%xsize]+1
        newposition.append((y,(x-1+xsize)%xsize))
        a[(y+1)%ysize,x]=a[(y+1)%ysize,x]+1
        newposition.append(((y+1)%ysize,x))
    
    if direction==2:
        a[y,(x-1+xsize)%xsize]=a[y,(x-1+xsize)%xsize]+1
        newposition.append((y,(x-1+xsize)%xsize))
        a[(y+1)%ysize,x]=a[(y+1)%ysize,x]+1
        newposition.append(((y+1)%ysize,x))
        a[y,(x+1)%xsize]=a[y,(x+1)%xsize]+1
        newposition.append((y,(x+1)%xsize))
                        
    if direction==3:
        a[(y+1)%ysize,x]=a[(y+1)%ysize,x]+1
        newposition.append(((y+1)%ysize,x))
        a[y,(x+1)%xsize]=a[y,(x+1)%xsize]+1
        newposition.append((y,(x+1)%xsize))
        a[(y-1+ysize)%ysize,x]=a[(y-1+ysize)%ysize,x]+1
        newposition.append(((y-1+ysize)%ysize,x))
    
    b=np.copy(a)
    return (b,newposition)      
```

```python
def explosion4(state,position):
    a=np.copy(state)
    (ysize,xsize)=np.shape(a)
    (y,x)=position
    newposition=[]
    
    a[y,x]=a[y,x]-4
    a[(y+1)%ysize,x]=a[(y+1)%ysize,x]+1
    newposition.append(((y+1)%ysize,x))
    
    a[y,(x+1)%xsize]=a[y,(x+1)%xsize]+1
    newposition.append((y,(x+1)%xsize))
        
    a[(y-1+ysize)%ysize,x]=a[(y-1+ysize)%ysize,x]+1
    newposition.append(((y-1+ysize)%ysize,x))
    
    a[y,(x-1+xsize)%xsize]=a[y,(x-1+xsize)%xsize]+1
    newposition.append((y,(x-1+xsize)%xsize))
                        
    b=np.copy(a)
    return (b,newposition)   
```

```python
def place(state,position,direction):
    a=np.copy(state)
    (ysize,xsize)=np.shape(a)
    (y,x)=position
    newposition=[]
    
    a[y,x]=a[y,x]-2
    if direction==0:
        a[(y+1)%ysize,x]=a[(y+1)%ysize,x]+2
     
    if direction==1:
        a[y,(x+1)%xsize]=a[y,(x+1)%xsize]+2
         
    if direction==2:
        a[(y-1+ysize)%ysize,x]=a[(y-1+ysize)%ysize,x]+2
                              
    if direction==3:
        a[y,(x-1+xsize)%xsize]=a[y,(x-1+xsize)%xsize]+2
           
    b=np.copy(a)
    return b      
```

```python
class parameter:
    def __init__(self,name,pStopRoll=0,pChangeDirection=0.2,pExp33=0.2,pExp330=1,pExp331=1,pExp44=0.1,pExp43=0.2,pExp42=0.3,\
                 pExp54=0.1,pExp53=0.2,pExp64=0.1,pExpMuti=0.01,pCap4=0,pCap5=0.3):
        self.name=name
        self.pStopRoll=pStopRoll
        self.pChangeDirection=pChangeDirection
        self.pExp33=pExp33
        self.pExp330=pExp330
        self.pExp331=pExp331
        self.pExp44=pExp44
        self.pExp43=pExp43
        self.pExp42=pExp42
        self.pExp54=pExp54
        self.pExp53=pExp53
        self.pExp64=pExp64
        self.pExpMuti=pExpMuti
        self.pCap4=pCap4
        self.pCap5=pCap5
    def change_pStopRoll(self,pSR):
        self.pStopRoll=pSR
    def change_pChangeDirection(self,pCD):
        self.pChangeDirection=pCD
    def change_pExp33(self,pE33):
        self.pExp33=pE33
    def change_pExp330(self,pE330):
        self.pExp330=pE330
    def change_pExp331(self,pE331):
        self.pExp330=pE331
    def change_pExp44(self,pE44):
        self.pExp44=pE44
    def change_pExp43(self,pE43):
        self.pExp43=pE43
    def change_pExp42(self,pE42):
        self.pExp42=pE42
    def change_pExp54(self,pE54):
        self.pExp54=pE54
    def change_pExp53(self,pE53):
        self.pExp53=pE53
    def change_pExp64(self,pE64):
        self.pExp64=pE64
    def change_pExpMuti(self,pEMuti):
        self.pExpMuti=pEMuti
    def change_pCap4(self,pC4):
        self.pCap4=pC4
    def change_pCap5(self,pC5):
        self.pCap5=pC5
```

```python
def experiment(initial_state,initial_operation,loop_number,parameters):
    ##### build the space to storage the states
    state_list=[]
    state_list.append(initial_state)       

    ##### build the space to storage the operations
    operation_list=[]
    operation_list.append(tuple(initial_operation))

    ##### loop the expriment
    loop_number=loop_number
    for loop in range(loop_number):

    ###### get current state    
        current_st=state_list[-1]
    #     print(current_st)
        next_st=np.copy(current_st)

    ###### get all the operation need to be done this time
        current_op=operation_list[-1]
        operation_number=len(current_op)
        next_op=[]

    ###### do rolling operation and record the next rolling's position and direction
        for i in range(operation_number):
            operation=current_op[i]
            if operation[0]==1:
    #             (next_st,newposition)=roll(next_st,operation[1],operation[2])
    #             next_op.append((1,newposition,operation[2]))

    ###### has pstop possibility to stop and (1-proll) possibility to keep rolling in the same direction            
                pstop=parameters.pStopRoll
                p=numpy.random.random()
                if p<pstop:
                    next_op.append((0,newposition,operation[2]))
                else:
                    proll=parameters.pChangeDirection
                    p=numpy.random.random()
                    if p>=proll:
                        (next_st,newposition)=roll(next_st,operation[1],operation[2])
                        next_op.append((1,newposition,operation[2]))
                    elif 0<=p<(0.5*proll):
                        (next_st,newposition)=roll(next_st,operation[1],operation[2])
                        next_op.append((1,newposition,(operation[2]+3)%4))
                    else:
                        (next_st,newposition)=roll(next_st,operation[1],operation[2])
                        next_op.append((1,newposition,(operation[2]+1)%4))

    ###### do explosion operation and record the out particles rolling's position and direction
            elif operation[0]==3:
                (next_st,newposition)=explosion3(next_st,operation[1],operation[2])
                next_op.append((1,newposition[0],(operation[2]+1)%4))
                next_op.append((1,newposition[1],(operation[2]+2)%4))
                next_op.append((1,newposition[2],(operation[2]+3)%4))

            elif operation[0]==2:
                (next_st,newposition)=explosion2(next_st,operation[1],operation[2])
                if operation[2]<=3:
                    next_op.append((1,newposition[0],(operation[2])%4))
                    next_op.append((1,newposition[1],(operation[2]+1)%4))
                else:
                    next_op.append((1,newposition[0],(operation[2])%4))
                    next_op.append((1,newposition[1],(operation[2]+2)%4))

            elif operation[0]==4:
                (next_st,newposition)=explosion4(next_st,operation[1])
                next_op.append((1,newposition[0],0))
                next_op.append((1,newposition[1],1))
                next_op.append((1,newposition[2],2))
                next_op.append((1,newposition[3],3))

            elif operation[0]==5:
                next_st=place(next_st,operation[1],operation[2])

    #####################################  all the operations are done already  ###################################################            
    ###### check if there is activated particles to explode        
        (ysize,xsize)=np.shape(next_st)       
        for j in range(ysize):
            for i in range(xsize):
                if next_st[j,i]==3:
    ###### (1-p3) possibility to release 3 particles                                   
                    p3=parameters.pExp33
                    p=numpy.random.random()
                    if p<p3:
    ###### hit make it more likely to explode in the opposite direction 
                        if current_st[j,i]==2:
                            p30=parameters.pExp330
                            p=numpy.random.random()
                            if p<p30:
                                if current_st[(j+1)%ysize,i]==1:
                                    next_op.append((3,(j,i),0))
                                elif current_st[j,(i+1)%xsize]==1:
                                    next_op.append((3,(j,i),1))
                                elif current_st[(j-1+ysize)%ysize,i]==1:
                                    next_op.append((3,(j,i),2))
                                elif current_st[j,(i-1+xsize)%xsize]==1:
                                    next_op.append((3,(j,i),3))
                            else:
                                dire=np.random.randint(4)
                                next_op.append((3,(j,i),dire))

                        elif current_st[j,i]==1:
                            p302=parameters.pExp331
                            p=numpy.random.random()
                            if p<p302:
                                if (current_st[(j+1)%ysize,i]==1)&(current_st[j,(i+1)%xsize]==1):
                                    p=numpy.random.random()
                                    if p>=0.5:
                                        next_op.append((3,(j,i),0))
                                    else:
                                        next_op.append((3,(j,i),1))

                                elif (current_st[j,(i+1)%xsize]==1)&(current_st[(j-1+ysize)%ysize,i]==1):
                                    p=numpy.random.random()
                                    if p>=0.5:
                                        next_op.append((3,(j,i),1))
                                    else:
                                        next_op.append((3,(j,i),2))

                                elif (current_st[(j-1+ysize)%ysize,i]==1)&(current_st[j,(i-1+xsize)%xsize]==1):
                                    p=numpy.random.random()
                                    if p>=0.5:
                                        next_op.append((3,(j,i),2))
                                    else:
                                        next_op.append((3,(j,i),3))

                                else:
                                    p=numpy.random.random()
                                    if p>=0.5:
                                        next_op.append((3,(j,i),3))
                                    else:
                                        next_op.append((3,(j,i),0))
                            else:
                                dire=np.random.randint(4)
                                next_op.append((3,(j,i),dire))

                        else:
                            dire=np.random.randint(4)
                            next_op.append((3,(j,i),dire))

                    else:
                        dire=np.random.randint(6)
                        next_op.append((2,(j,i),dire))

    ###### 4 particles situation             
                if next_st[j,i]==4:
                    p40=parameters.pExp44
                    p41=parameters.pExp43
                    p42=parameters.pExp42
                    p=numpy.random.random()
                    if p<p40:
                        next_op.append((4,(j,i)))
                    elif p41+p40>p>=p40:
                        dire=np.random.randint(4)
                        next_op.append((3,(j,i),dire))
                    elif p42+p41+p40>p>=p41+p40:
                        dire=np.random.randint(6)
                        next_op.append((2,(j,i),dire))
                    else:
                        if next_st[(j+1)%ysize,i]==0:
                            next_op.append((5,(j,i),0))
                        elif next_st[j,(i+1)%xsize]==0:
                            next_op.append((5,(j,i),1))
                        elif next_st[(j-1+ysize)%ysize,i]==0:
                            next_op.append((5,(j,i),2))
                        else:
                            next_op.append((5,(j,i),3))

    ###### 5 particles situation             
                if next_st[j,i]==5:
                    p51=parameters.pExp54
                    p52=parameters.pExp53
                    p=numpy.random.random()
                    if p<p51:
                        next_op.append((4,(j,i)))
                    elif p52+p51>p>=p51:
                        dire=np.random.randint(4)
                        next_op.append((3,(j,i),dire)) 
                    else:
                        if next_st[(j+1)%ysize,i]==0:
                            next_op.append((5,(j,i),0))
                        elif next_st[j,(i+1)%xsize]==0:
                            next_op.append((5,(j,i),1))
                        elif next_st[(j-1+ysize)%ysize,i]==0:
                            next_op.append((5,(j,i),2))
                        else:
                            next_op.append((5,(j,i),3))

    ###### 6 particles situation                  
                if next_st[j,i]==6:
                    p6=parameters.pExp64
                    p=numpy.random.random()
                    if p<p6:
                        next_op.append((4,(j,i)))
                    else:
                        if next_st[(j+1)%ysize,i]==0:
                            next_op.append((5,(j,i),0))
                        elif next_st[j,(i+1)%xsize]==0:
                            next_op.append((5,(j,i),1))
                        elif next_st[(j-1+ysize)%ysize,i]==0:
                            next_op.append((5,(j,i),2))
                        else:
                            next_op.append((5,(j,i),3))

    ###### mutiple particles situation 
                if next_st[j,i]>6:
                    pm=parameters.pExpMuti
                    p=numpy.random.random()
                    if p<pm:
                        next_op.append((4,(j,i)))
                    else:
                        if next_st[(j+1)%ysize,i]==0:
                            next_op.append((5,(j,i),0))
                        elif next_st[j,(i+1)%xsize]==0:
                            next_op.append((5,(j,i),1))
                        elif next_st[(j-1+ysize)%ysize,i]==0:
                            next_op.append((5,(j,i),2))
                        else:
                            next_op.append((5,(j,i),3))

    ###### check if the rolling in the next frame can be done
        check_op=next_op.copy()
        for g in range(len(check_op)):
            operation=check_op[g]
            if operation[0]==1:
                (j,i)=operation[1]

    ##### only sigle parcile can move
                if next_st[j,i]!=1:
                    next_op[g]=(0,0,0)

    ##### captured by the 'environment'
                elif (next_st[(j+1)%ysize,i]+next_st[j,(i+1)%xsize]+next_st[(j-1+ysize)%ysize,i]+next_st[j,(i-1+xsize)%xsize])==4:
                    pcap4=parameters.pCap4
                    p=numpy.random.random()
                    if p<=pcap4:
                        next_op[g]=(0,0,0)
                elif (next_st[(j+1)%ysize,i]+next_st[j,(i+1)%xsize]+next_st[(j-1+ysize)%ysize,i]+next_st[j,(i-1+xsize)%xsize])==5:
                    pcap5=parameters.pCap5
                    p=numpy.random.random()
                    if p<=pcap5:
                        next_op[g]=(0,0,0)
                elif (next_st[(j+1)%ysize,i]+next_st[j,(i+1)%xsize]+next_st[(j-1+ysize)%ysize,i]+next_st[j,(i-1+xsize)%xsize])>=6:
                    next_op[g]=(0,0,0)

    ###### check if any error
        for j in range(ysize):
            for i in range(xsize):
                if next_st[j,i]<0:
                    print('error')
                    print(loop)

        operation_list.append(tuple(next_op))
        state_list.append(next_st) 
        
    return(state_list,operation_list)
```
